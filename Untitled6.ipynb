```python
### ATIVIDADE
## 1-Calculando bônus dos funcionários
#Uma empresa deseja calcular o bônus de seus funcionários com base nas vendas realizadas no mês. O bônus será concedido apenas para aqueles que atingirem ou ultrapassarem a meta de R$ 3.000,00. O valor do bônus
#corresponde a 5% do total de vendas do funcionário. Dado a lista de vendas dos funcionários:

vendas_funcionarios = [2800, 3200, 1500, 4000, 5000, 1000, 3500]
meta = 3000
porcentual_bonus = 0.05
for venda in vendas_funcionarios:
    if venda >= meta:
      bonus = porcentual_bonus * venda
    else:
      bonus = 0
    print(bonus)

```

```python
### ATIVIDADE
## 2-Contagem de produtos vendidos acima da meta
#Uma loja de eletrônicos quer contar quantos dias tiveram vendas acima da meta de R$ 2.500,00. O gerente forneceu a lista com os valores das vendas diárias da semana:

vendas_diarias = [2000, 3000, 2500, 4000, 1800, 5000, 2700]
meta = 30
dias_acima_da_meta = 0

for venda in vendas_diarias:
    if venda > meta:
      dias_acima_da_meta += 1

print(dias_acima_da_meta)
```

```python
###ATIVIDADE
## 3-Cálculo de comissão dos vendedores
# Uma empresa paga uma comissão de 8% sobre as vendas para os vendedores que superarem a meta de R$2.000,00. Dada a lista de vendas realizadas por cada vendedor:

vendas_vendedores = [1500, 2500, 1800, 3200, 4000, 1200, 2800]
meta = 2000
comissão = 0.08
for venda in vendas_vendedores:
    if venda >= meta:
      bonus = comissão * venda
    else:
      bonus = 0
    print(bonus)
```

```python
### ATIVIDADE
## 4-Contando vendas que atingiram a meta
# Dada a lista de vendas:

lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
contador = 0

for venda in lista_vendas:
  if venda > meta:
   contador += 1
print("Quantidade de vendas que atingiram a meta", contador)
```

```python
### ATIVIDADE
##5-Aplicando um bônus extra para grandes vendas
#Além do bônus de 10% para vendas que atingem ou superam R$ 1200,00, vendas acima de R$ 2000,00 recebem um bônus extra de R$ 50,00. Dada a lista de vendas:

lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_bonus = 0.1
bonus_extra = 50

for venda in lista_vendas:
    bonus = 0

    if venda >= meta:
        bonus = venda * percentual_bonus

        if venda >= 2000:
            bonus = bonus + bonus_extra
    print(f"venda: {venda} -> bonus : R${bonus}")
```

```python
### ATIVIDADE
## 6-Encontrando a maior venda realizada
# Dada a lista de vendas abaixo:

lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

maior_vendas = lista_vendas[0]

for venda in lista_vendas:
  if venda > maior_vendas:
    maior_vendas = venda
print("A maior venda", maior_vendas)
```
